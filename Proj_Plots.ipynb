```python
os.getcwd()
os.chdir(r'C:\Users\sherw\OneDrive\Desktop\outputs\Cartpole_FF')

path_dd = 'ddqn_scalars.pkl'
f_dd = open(path_dd, 'rb')
dd = pickle.load(f_dd)
f_dd.close()

path_td = 'tdqn_scalars.pkl'
f_td = open(path_td, 'rb')
td = pickle.load(f_td)
f_td.close()

path_sd = 'sd_scalars.pkl'
f_sd = open(path_sd, 'rb')
sd = pickle.load(f_sd)
f_sd.close()

path_sd = 'fd_scalars.pkl'
f_sd = open(path_sd, 'rb')
fd = pickle.load(f_sd)
f_sd.close()
```
